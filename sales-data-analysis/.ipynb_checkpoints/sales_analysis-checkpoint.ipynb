## 3. Veri Temizleme

```python
df=df.drop(columns=["ORDERNUMBER", "PHONE", "ADDRESSLINE2", "ADDRESSLINE1", "STATE", "TERRITORY", "POSTALCODE", "CONTACTFIRSTNAME", "CONTACTLASTNAME", "ORDERDATE"])
df.isnull().sum()
```

## 4. Analiz ve Görselleştirme

```python
sales_for_countries=df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False)
```

```python
# ülkelere göre satış miktarlarını veren grafik.
sales_for_countries.plot(kind="bar", xlabel="Ülkeler", ylabel="Satışlar", figsize=(10,8), title="Ülkeler vs Satış Miktarları")
```

```python
sales_for_product_lines=df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)
# ürün kategorilerine göre satışları veren grafik.
sales_for_product_lines.plot(kind="bar", xlabel="ÜRÜN KATEGORİLERİ", ylabel="SATIŞ MİKTARLARI", title="Ürün Kategorilerine Göre Satışlar")
```

```python
sales_for_years=df.groupby("YEAR_ID")["SALES"].sum()
# yıllara göre satış miktarlarını veren grafik.
sales_for_years.plot(kind="line", xlabel="Yıllar", ylabel="Satış Miktarları", title="Yıllar vs Satış Miktarları", figsize=(10,8), grid=True)
```

```python
sales_for_months=df.groupby("MONTH_ID")["SALES"].sum().sort_values(ascending=False)
# aylara göre satış miktarlarını veren grafik.
sales_for_months.plot(kind="bar", xlabel="Aylar", ylabel="Satış Miktarları", title="Aylar vs Satış Miktarları", figsize=(10,8), grid=True)
```

```python
sales_for_deal_sizes=df.groupby("DEALSIZE")["SALES"].sum()
# dealsize sütununun dağılımını gösteren pasta grafiği.
sales_for_deal_sizes.plot(kind="pie")
```

```python
df.groupby(["COUNTRY", "PRODUCTLINE"])["SALES"].sum().head(10)
# en çok satılan ürünler ve bu ürünlerin satıldığı ülkeler.
```
